```python
from gensim.models.phrases import Phrases, Phraser
from nltk.stem import SnowballStemmer
#nltk.download('wordnet')
#nltk.download('stopwords')
import re
```

```python
def clean_corpus(corpus, threshold=50):
    # Word Tokenization
    tokenized_words = []
    for sentence in corpus:
        # remove special characters such as parentheses and commas
        sentence = re.sub(r'[^\w\s]', '', sentence)
        sentence_words = nltk.tokenize.word_tokenize(sentence)
        tokenized_words.append(sentence_words)

    # Stopwords
    stopwords = nltk.corpus.stopwords.words('french')
    tokenized_words = [[word for word in words if word.lower() not in stopwords] for words in tokenized_words]

    # Stemming
    #stemmer = SnowballStemmer('french')
    #tokenized_words = [[stemmer.stem(word) for word in words] for words in tokenized_words]

    # Lemmatization (à la place de stemming, tester la différence)
    lemmatizer = nltk.stem.WordNetLemmatizer()
    tokenized_words = [[lemmatizer.lemmatize(word) for word in words] for words in tokenized_words]

    # Phrases
    phrases = Phrases(min_count=1, threshold=threshold)
    phrases.add_vocab(tokenized_words)
    phraser = Phraser(phrases)

    # Merging multi-word expressions in the tokenization
    clean_corpus = []
    for sentence in tokenized_words:
        sentence_phrases = phraser[sentence]
        clean_corpus.append(sentence_phrases)

    return clean_corpus
```
